```python
import pandas as pd
import numpy as np

# Step 1: Load data, treat "?" as missing
df = pd.read_csv("chronic_kidney_disease_full.csv", na_values='?')

# Step 2: Clean column names (remove quotes and whitespace)
df.columns = [col.strip().strip("'").strip('"') for col in df.columns]

# Step 3: Convert numerics where possible
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')

# Step 4: Impute missing values
for col in df.columns:
    if df[col].dtype == 'object':
        # Categorical column â€” fill with mode
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        # Numeric column â€” fill with mean
        df[col].fillna(df[col].mean(), inplace=True)

# Step 5: Optional â€” check final result
print(df.isnull().sum())  # Should show 0 for all columns

# Save clean version
df.to_csv("cleaned_chronic_kidney.csv", index=False)
```

```python
import csv

columns = ['age','bp','sg','al','su','rbc','pc','pcc','ba','bgr','bu','sc','sod','pot',
           'hemo','pcv','wbcc','rbcc','htn','dm','cad','appet','pe','ane','class']

ckd = {}
nckd = {}

with open('completedata.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    linecount = 0
    for row in csv_reader:
        if linecount == 0:
            ckd = row
        else:
            nckd = row
        linecount = linecount + 1

# Round and format representative values
ckd["age"] = str(round(float(ckd["age"])))
ckd["bp"] = str(round(float(ckd["bp"])))
ckd["al"] = str(round(float(ckd["al"])))
ckd["su"] = str(round(float(ckd["su"])))
ckd["bgr"] = str(round(float(ckd["bgr"])))
ckd["bu"] = str(round(float(ckd["bu"])))
ckd["sc"] = str(round(float(ckd["sc"])))
ckd["wbcc"] = str(round(float(ckd["wbcc"])))
ckd["pcv"] = str(round(float(ckd["pcv"])))
ckd["sod"] = str(round(float(ckd["sod"])))

nckd["age"] = str(round(float(nckd["age"])))
nckd["bp"] = str(round(float(nckd["bp"])))
nckd["al"] = str(round(float(nckd["al"])))
nckd["su"] = str(round(float(nckd["su"])))
nckd["bgr"] = str(round(float(nckd["bgr"])))
nckd["bu"] = str(round(float(nckd["bu"])))
nckd["sc"] = str(round(float(nckd["sc"])))
nckd["wbcc"] = str(round(float(nckd["wbcc"])))
nckd["pcv"] = str(round(float(nckd["pcv"])))
nckd["sod"] = str(round(float(nckd["sod"])))
print(ckd)
print(nckd)

writelist = []

with open('cleaned_chronic_kidney.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    raw_headers = next(csv_reader)
    headers = [h.strip().strip("'\"") for h in raw_headers]  # Clean header names
    csv_reader = csv.DictReader(csv_file, fieldnames=headers)
    next(csv_reader)
    for row in csv_reader:
        for attrib in columns:
            if row[attrib] == "?":
                if row["class"] == "ckd":
                    row[attrib] = ckd[attrib]
                else:
                    row[attrib] = nckd[attrib]
            else:
                continue
        writelist.append(row)

with open('preprocessed.csv', mode='w', newline='') as csv_file:
    writer = csv.DictWriter(csv_file, columns)
    writer.writeheader()
    for row in writelist:
        try:
            writer.writerow(row)
        except:
            print(row)
```

```python
import csv

columns = ['age', 'bp', 'sg', 'al', 'su', 'rbc', 'pc', 'pcc', 'ba', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'class']
writerows = []

with open('preprocessed.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        for attrib in columns:
            if row[attrib] == "normal" or row[attrib] == "present":
                row[attrib] = 1
            elif row[attrib] == "yes" or row[attrib] == "good" or row[attrib] == "ckd":
                row[attrib] = 1
            elif row[attrib] == "abnormal" or row[attrib] == "notpresent":
                row[attrib] = 0
            elif row[attrib] == "no" or row[attrib] == "poor" or row[attrib] == "notckd":
                row[attrib] = 0
        writerows.append(row)

with open('labelencoded.csv', mode='w', newline='') as csv_file:
    writer = csv.DictWriter(csv_file, columns)
    writer.writeheader()
    for row in writerows:
        try:
            writer.writerow(row)
        except:
            print(row)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

df = pd.read_csv('labelencoded.csv')
df.drop(["rbc", "rbcc", "wbcc"], axis=1, inplace=True)

x = df.drop('class', axis=1)
y = df['class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=13)

classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)
print(y_pred)

accuracy = accuracy_score(y_test, y_pred) * 100
print(accuracy)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Load CSV
df = pd.read_csv('labelencoded.csv')

# Strip leading/trailing whitespace from all string cells
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Replace common invalid placeholders with NaN
df.replace(['?', '\t?', '', ' '], np.nan, inplace=True)

# Drop rows with any NaN values
df.dropna(inplace=True)

# Drop unnecessary columns
df.drop(["rbc", "rbcc", "wbcc"], axis=1, inplace=True)

# Split into features and target
x = df.drop('class', axis=1)
y = df['class']

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=13)

# Train Gaussian Naive Bayes
model = GaussianNB()
model.fit(x_train, y_train)

# Predict and evaluate
y_pred = model.predict(x_test)
print("Predictions:", y_pred)

accuracy = accuracy_score(y_test, y_pred) * 100
print("Accuracy is:", accuracy)
```
